```python
#load dragonn tutorial utilities 
%reload_ext autoreload
%autoreload 2
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import pybedtools
```

```python
from keras.models import load_model
from kerasAC.metrics import recall, specificity, fpr, fnr, precision, f1
from kerasAC.custom_losses import ambig_binary_crossentropy
import pickle 
from kerasAC.generators import * 
from scipy.special import logit 
```

```python
gwas_pd = pd.read_csv('/mnt/lab_data3/soumyak/adpd/gwas/23andme_PD.txt.gz', sep='\t')
gwas_pd.rename(columns={'snp_pos':'start'}, inplace=True)
gwas_pd['end'] = gwas_pd['start'] + 1
gwas_pd = gwas_pd[['chr', 'start', 'end', 'ref', 'alt', 'rsid', 'pvalue']]
gwas_pd.head()
```

```python
gwas_pd_filtered = gwas_pd[gwas_pd['pvalue'] < 5e-8]
gwas_pd_filtered.head()
```

```python
gwas_pd_filtered_mapt = gwas_pd_filtered.loc[(gwas_pd_filtered['chr'] == 'chr17') & (gwas_pd_filtered['start'] >= 44890000) & (gwas_pd_filtered['end'] <= 46390000)]
gwas_pd_filtered_mapt.head()
```

```python
gwas_ad = pd.read_csv('/mnt/lab_data3/soumyak/adpd/gwas/GWAS_Kunkle2019.txt.gz', sep='\t')
gwas_ad.rename(columns={'snp_pos':'start'}, inplace=True)
gwas_ad['end'] = gwas_ad['start'] + 1
gwas_ad['chr'] = gwas_ad['chr'].apply(lambda x: 'chr' + str(x))
gwas_ad = gwas_ad[['chr', 'start', 'end', 'non_effect_allele', 'effect_allele', 'rsid', 'pvalue']]
gwas_ad.head()
```

```python
gwas_ad_filtered = gwas_ad[gwas_ad['pvalue'] < 5e-8]
gwas_ad_filtered.head()
```

```python
gwas_ad_filtered_mapt = gwas_ad_filtered.loc[(gwas_ad_filtered['chr'] == 'chr17') & (gwas_ad_filtered['start'] >= 44890000) & (gwas_ad_filtered['end'] <= 46390000)]
gwas_ad_filtered_mapt.head()
```

```python
gwas_pd_best = gwas_pd_filtered.sort_values('pvalue')
gwas_pd_best.head()
```

```python
dict_init = {'chr':[], 'start':[], 'end':[], 'ref':[], 'alt':[], 'rsid':[], 'pvalue':[]}
gwas_pd_best_x = pd.DataFrame(dict_init)
#gwas_pd_best_x = gwas_pd_best.iloc[[0], :]
x = 5
for ind,row in gwas_pd_best.iterrows():
    if len(gwas_pd_best_x) > x:
        break
    else:
        new = True
        for ind2,row2 in gwas_pd_best_x.iterrows():
            if row2['start'] - 50000 <  row['start'] and row2['end'] + 50000 > row['end'] and row2['chr'] == row['chr']:
                new = False
        if new and row['chr'] in ['chr1', 'chr19', 'chr2', 'chr3', 'chr20']:
            dict_ = {'chr':[], 'start':[], 'end':[], 'ref':[], 'alt':[], 'rsid':[], 'pvalue':[]}
            for i in row.keys():
                dict_[i].append(row[i])
            new_df = pd.DataFrame(dict_)
            gwas_pd_best_x = gwas_pd_best_x.append(new_df, ignore_index=True)
gwas_pd_best_x           
```

```python
gwas_ad_best = gwas_ad_filtered.sort_values('pvalue')
gwas_ad_best
```

```python
dict_init = {'chr':[], 'start':[], 'end':[], 'non_effect_allele':[], 'effect_allele':[], 'rsid':[], 'pvalue':[]}
gwas_ad_best_x = pd.DataFrame(dict_init)                                        
#gwas_ad_best_x = gwas_ad_best.iloc[[0], :]                                     
x = 5                                                                          
for ind,row in gwas_ad_best.iterrows():                                         
    if len(gwas_ad_best_x) > x:                                                 
        break                                                                   
    else:
        new = True                                                              
        for ind2,row2 in gwas_ad_best_x.iterrows():
            if row2['start'] - 50000 <  row['start'] and row2['end'] + 50000 > row['end'] and row2['chr'] == row['chr']:
                new = False                                                     
        if new and row['chr'] in ['chr1', 'chr19', 'chr2', 'chr3', 'chr20']:
            dict_ = {'chr':[], 'start':[], 'end':[], 'non_effect_allele':[], 'effect_allele':[], 'rsid':[], 'pvalue':[]}
            for i in row.keys():                                                
                dict_[i].append(row[i])                                         
            new_df = pd.DataFrame(dict_)                                        
            gwas_ad_best_x = gwas_ad_best_x.append(new_df, ignore_index=True)   
gwas_ad_best_x
```

```python
gwas_pd_ld = gwas_pd_best_x
gwas_pd_ld['start'] = gwas_pd_ld['start'] - 50000
gwas_pd_ld['end'] = gwas_pd_ld['end'] + 50000
counter = 0
gwas_pd_bed = pybedtools.BedTool.from_dataframe(gwas_pd)
for ind, row in gwas_pd_ld.iterrows():
    counter += 1
    dict_ = {'chr':[], 'start':[], 'end':[], 'ref':[], 'alt':[], 'rsid':[], 'pvalue':[]}
    for i in row.keys():                                                
        if i == 'start' or i == 'end':
            dict_[i].append(int(row[i]))
        else:
            dict_[i].append(row[i])
    new_df = pd.DataFrame(dict_)
    print(new_df)
    print(gwas_pd)
    gwas_pd_ld_bed = pybedtools.BedTool.from_dataframe(new_df)
    gwas_pd_intersect_bed = gwas_pd_bed.intersect(gwas_pd_ld_bed, u=True)
    gwas_pd_intersect_bed.saveas('gwas_pd_intersect_bed_'+str(counter)+'.txt')
```

```python
gwas_ad_ld = gwas_ad_best_x                                                     
gwas_ad_ld['start'] = gwas_ad_ld['start'] - 50000                               
gwas_ad_ld['end'] = gwas_ad_ld['end'] + 50000                                   
counter = 0                                                                     
gwas_ad_bed = pybedtools.BedTool.from_dataframe(gwas_ad)
for ind, row in gwas_ad_ld.iterrows():                                          
    counter += 1                                                                
    dict_ = {'chr':[], 'start':[], 'end':[], 'non_effect_allele':[], 'effect_allele':[], 'rsid':[], 'pvalue':[]}
    for i in row.keys():                                                
        if i == 'start' or i == 'end':
            dict_[i].append(int(row[i]))
        else:
            dict_[i].append(row[i])
    new_df = pd.DataFrame(dict_)
    print(new_df)
    print(gwas_ad)                                                
    gwas_ad_ld_bed = pybedtools.BedTool.from_dataframe(new_df)                                      
    gwas_ad_intersect_bed = gwas_ad_bed.intersect(gwas_ad_ld_bed, u=True)       
    gwas_ad_intersect_bed.saveas('gwas_ad_intersect_bed_'+str(counter)+'.txt')
```

```python
##load the model 
import pickle
from kerasAC.metrics import tpr, tnr, fpr, fnr, precision, f1, recall
from kerasAC.custom_losses import ambig_binary_crossentropy, ambig_mean_squared_error
import numpy as np
fold = 0
custom_objects={"sensitivity":tpr,
                "specificity":tnr,
                "fpr":fpr,
                "fnr":fnr,
                "recall":recall,
                "precision":precision,
                "f1":f1,
                "ambig_binary_crossentropy":ambig_binary_crossentropy,
                "ambig_mean_squared_error":ambig_mean_squared_error}

model=load_model("/mnt/lab_data3/soumyak/adpd/deeplearning/models/regression/PD_CTRL_MDTG.model."+str(fold),custom_objects=custom_objects)
```

```python
# get model predictions for bQTLs 
from dragonn.generators import * 
bqtl_post_gen=BQTLGenerator("gwas_pd_intersect_bed_1.txt","/mnt/data/annotations/by_release/hg38/GRCh38_no_alt_analysis_set_GCA_000001405.15.fasta.gz","ref")
bqtl_alt_gen=BQTLGenerator("gwas_pd_intersect_bed_1.txt","/mnt/data/annotations/by_release/hg38/GRCh38_no_alt_analysis_set_GCA_000001405.15.fasta.gz","alt")
```

```python
ref_predictions=model.predict_generator(bqtl_post_gen,
                                                                    max_queue_size=5000,
                                                                        workers=40,
                                                                          use_multiprocessing=True,
                                                                          verbose=1)
```

```python
alt_predictions=model.predict_generator(bqtl_alt_gen,
                                                               max_queue_size=5000, 
                                                               workers=40, 
                                                               use_multiprocessing=True, 
                                                               verbose=1)


```
